```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned.csv')
df
```

```python
df_copy = df.copy()
df_copy.loc[(df['tasks'] == 'object_detection'),'datasets'] = 1
df_copy
```

```python
all_record = {}
da = pd.read_csv('stat.csv') 
da_to = da.copy()
da_ta = da.copy()
da_da = da.copy()
total = []
task_li = []
data_li = []
count=0
for index, row in df.iterrows():
    
    area = [x.strip() for x in row["multi"].split(',') if x][0]

    task = [x.strip() for x in row["tasks"].split(',') if x]

    dataset = [x.strip() for x in row["datasets"].split(',') if x]
    if ('image' in dataset) and (len(dataset) >= 2):
        dataset.remove('image')
    elif ('text' in dataset) and (len(dataset) >= 2):
        dataset.remove('text')
    
    try:
        model = [x.strip() for x in row["models"].split(',') if x]
    except AttributeError:
        model = []
        
    record = [task,dataset,model]
    all_record[index] = record

    for i in task:
        iflag = 0
        if i == 'nlp' or i == 'cv':
            continue
        for i_item in task_li:
            if area == i_item['class'] and i == i_item['task']:
                i_item['count'] += 1
                iflag = 1
                break
        if iflag != 1:
            new_i_item = {}
            new_i_item['class'] = area
            new_i_item['task'] = i
            new_i_item['count'] = 1
            task_li.append(new_i_item)
        for j in dataset:
            jflag = 0
            for j_item in data_li:
                if area == j_item['class'] and i == j_item['task'] and j == j_item['dataset']:
                    j_item['count'] += 1
                    jflag = 1
                    break
            if jflag != 1:
                new_j_item = {}
                new_j_item['class'] = area
                new_j_item['task'] = i
                new_j_item['dataset'] = j
                new_j_item['count'] = 1
                data_li.append(new_j_item)
            for k in model:
                flag = 0
                for item in total:
                    if area == item['class'] and i.lower() == item['task'].lower() and j.lower() == item['dataset'].lower() and k.lower() == item['model'].lower():
                        item['count'] += 1
                        flag = 1
                        break
                if flag != 1:
                    new_item = {}
                    new_item['class'] = area
                    new_item['task'] = i.lower()
                    new_item['dataset'] = j.lower()
                    new_item['model'] = k.lower()
                    new_item['count'] = 1
                    total.append(new_item)
```

```python
for item in total:
    da_to = da_to.append(item, ignore_index=True)
for item in task_li:
    da_ta = da_ta.append(item, ignore_index=True)
for item in data_li:
    da_da = da_da.append(item, ignore_index=True)
```

```python
da_ta
```

```python
plt.figure(figsize=(20,8),dpi=200)
class_dic = da_ta['count'].groupby(da_ta['class']).sum().to_dict()
class_li = da_ta['count'].groupby(da_ta['class']).sum().tolist()
class_dic.keys()
plt.title('how complex the tasks is')
label = list(class_dic.keys())
plt.bar([i for i in range(len(class_li))],class_li,tick_label=label)
plt.xticks(rotation=80)
plt.xlabel('class')
plt.ylabel('count')
plt.savefig('tasks')
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=200)
nlp = da_ta.loc[(da_ta['class'] == 'nlp') | (da_ta['class'] == 'nlp-multi')]
nlp_co = nlp.copy()
nlp_task = nlp_co['count'].groupby(nlp_co['task']).sum().to_dict()
nlp_task_li = nlp_co['count'].groupby(nlp_co['task']).sum().tolist()
plt.title('nlp tasks')
label = list(nlp_task.keys())
plt.bar([i for i in range(len(nlp_task_li))],nlp_task_li,tick_label=label)
plt.xticks(rotation=80)
plt.xlabel('tasks')
plt.ylabel('count')
plt.savefig('nlp_tasks.png')
plt.show()
```
